# [dbo].[v_clientes]

```sql
/****** Object:  View [dbo].[v_clientes]    Script Date: 10/02/2023 17:38:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
---se creara una vista para los datos

create view [dbo].[v_clientes]
as

select 
 cl.ClienteId
,cl.Nombre
,cl.Telefono
,cl.Correo
,cl.edad
,cl.MontoSolicitud
,cl.Estatus
,cl.Aprobación
,cl.FechaAlta
,pa.MontoPagado
,pa.FechaPago

from Clientes cl
inner join Pagos pa
on cl.ClienteId = pa.ClienteId;









GO
```
